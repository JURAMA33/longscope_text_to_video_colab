```python
import os
if not os.path.exists('/content/Text-To-Video-Finetuning'):
    %cd /content
    !git clone -b dev https://github.com/camenduru/Text-To-Video-Finetuning

%cd /content/Text-To-Video-Finetuning
prompt = "waterfall" #@param {type:"string"}
negative = "text, watermark, copyright, blurry, low resolution, blur, low quality" #@param {type:"string"}
prompt_q = f"\"{prompt}\""
negative_q = f"\"{negative}\""
num_steps = 25 #@param {type:"raw"}
guidance_scale = 23 #@param {type:"raw"}
fps = 24 #@param {type:"raw"}
num_frames = 96 #@param {type:"raw"}
!python inference.py -m "/content/model" -p {prompt_q} -n {negative_q} -W 384 -H 216 -o /content/outputs -d cuda -x -s {num_steps} -g {guidance_scale} -f {fps} -T {num_frames}
```

```python
import glob
import os
from IPython.display import HTML
from base64 import b64encode

# Buscar el video generado más recientemente
video_files = glob.glob('/content/outputs/*.mp4')
if video_files:
    latest_video = max(video_files, key=os.path.getctime)
    print(f'Mostrando video: {latest_video}')

    # Leer y codificar el video para mostrarlo en el notebook
    mp4 = open(latest_video, 'rb').read()
    data_url = "data:video/mp4;base64," + b64encode(mp4).decode()

    display(HTML(f"""
    <video width=640 controls>
          <source src="{data_url}" type="video/mp4">
    </video>
    """))
else:
    print('No se encontraron videos en /content/outputs. Por favor, asegúrate de que la celda de generación se ejecutó correctamente.')
```
